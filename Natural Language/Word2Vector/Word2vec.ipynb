```python
from gensim.models.keyedvectors import KeyedVectors
```

```python
word_vectors=KeyedVectors.load_word2vec_format('../../Data/GoogleNews-vectors-negative300.bin.gz', binary=True)
len(word_vectors.vocab)
```

```python
word_vectors.most_similar(positive=['germany', 'france'], topn=5)
```

```python
 word_vectors.doesnt_match("python javascript people computer lying".split())
```

```python
word_vectors.most_similar(positive=['Time', 'travel', 'world'], negative=[], topn=5)
```
