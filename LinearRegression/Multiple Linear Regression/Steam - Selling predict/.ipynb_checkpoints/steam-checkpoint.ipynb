## Correlation

## $$ \rho _{XY} = corr(X,Y)$$
## $$ -1.0 \leq \rho _{XY} \leq +1.0 $$

```python
mask = np.zeros_like(data.corr())
triangle_indices = np.triu_indices_from(mask)
mask[triangle_indices] = True
```

```python
# Correlation analasys
plt.figure(figsize=(16,10))
sn.heatmap(data.corr(), mask=mask, annot=True, annot_kws={"size": 14})
sn.set_style('white')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show() # not so strong
```

```python
#Onother way
from pandas.plotting import scatter_matrix
attributes = ["price", "english", "required_age", "achievements"]
scatter_matrix(data[attributes], figsize=(12, 8))

data.plot(kind="scatter", x="price", y="achievements", alpha=0.1) #not so strong
```

```python
# CLEANING
data=data.drop(columns=["release_date","name","developer","appid","english","achievements"])
data=data[data.price <= 55]
data.shape

le = preprocessing.LabelEncoder()
data.genres=le.fit_transform(data.genres)
data.steamspy_tags=le.fit_transform(data.steamspy_tags)
data.categories=le.fit_transform(data.categories)
data.publisher=le.fit_transform(data.publisher)
data.platforms=le.fit_transform(data.platforms)

data
```

# Training and Evaluating on the Training Set

```python
X=data.loc[:, data.columns != 'owners']
y=data.owners

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, reg.predict(X_test))
```
